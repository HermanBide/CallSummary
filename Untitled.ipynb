```python
def getStationData(fromstation, tostation):
    browser = RoboBrowser(history=True)
    br = RoboBrowser()
    br.open("http://www.delhimetrorail.com/metro-fares.aspx")
    form = br.get_form()
    form['ctl00$MainContent$ddlFrom'].value = fromstation
    form['ctl00$MainContent$ddlTo'].value = tostation
    br.submit_form(form)
    return str(br.parsed())
```

```python
def getStationFare ( fro , to , url ):
    br = RoboBrowser ( history=True , parser="html.parser" )
    br.open ( url )
    form = br.get_form ( )
    form[ 'ctl00$MainContent$ddlFrom' ].value = str ( fro )
    form[ 'ctl00$MainContent$ddlTo' ].value = str ( to )
    br.submit_form ( form )
    src = str ( br.parsed ( ) )
    return src
```
